After everything is said and done, you can have fun and just generate some morphologies!

```python

import numpy as np
import torch
from gan.network import dc_generator
from gan.network_utils import gan_generate
from gan.network_utils import load_model
from gan.network_utils import show_images

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline

%load_ext autoreload
%autoreload 2

plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'


```

```python
# Load in models

batch_size = 64
dtype = torch.cuda.FloatTensor
directory_granule = 'models/granule/'
directory_pyramidal = 'models/pyramidal/'

G_granule = dc_generator(batch_size).type(dtype)
G_pyramidal = dc_generator(batch_size).type(dtype)

load_model(G_granule, directory_granule + 'trained-granule-generator-ls-1-in-1.pt')
load_model(G_pyramidal, directory_pyramidal + 'trained-pyramidal-generator-ls-1-in-0.pt')

print('models loaded')
```

```python
# Generate granule dendrite morphologies

fake_granule = gan_generate(dtype, G_granule)
show_images(fake_granule)
```

```python
# Generate pyramidal morphologies

fake_pyramidal = gan_generate(dtype, G_pyramidal)
show_images(fake_pyramidal)
```
